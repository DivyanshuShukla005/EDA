```python
concode=pd.read_excel("Country-Code.xlsx")
```

```python
concode.head(2)
```

```python
crux=pd.merge(zomato,concode,on="Country Code",how="left")
# merging both table by country code(treated as primary key)is in both table
# putting left join (watch how)
# this is our final table 
#apply join according to need . 
```

```python
crux.Country.value_counts()
# In this code we get country with name with values 
# im going to show in pie chart so segregate name and value in different variable 
```

```python
country_names=crux.Country.value_counts().index
# we stored country name in this variable
```

```python
country_val=crux.Country.value_counts().values
# we store values  in this variable
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
# their are so many country so, we taake top 3 country ([:3])
# autopct stands for the percentage ('%1.2f%%'= means after decimal 2 number will come.)
```

Observation : Zomato maximum record or transaction ar from india after that USA and then united kingdoms

```python
crux.groupby(['Aggregate rating','Rating color','Rating text']).size()
# make a dataframe by doing group by function
# with size we see all the data which are store in these variables
```

```python
ratings=crux.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Ratings Counts"})
# now we are resetting the index & renaming it like (where their is 0 in column name so,it will make Rating counts)
# we groupby all variables so, last rating counts column occurs it means (we have 2148 records for this => 0.0,white,not rated )
# at last we keep all this information into rating variable
```

```python
ratings
```

## Observation
When Rating is between 4.5 to 4.9---> Excellent

When Rating are between 4.0 to 4.4--->very good

when Rating is between 3.5 to 3.9----> good

when Rating is between 2.5 to 3.4----> average

when Rating is between 1.8 to 2.4----> Poor

```python
ratings.head(2)
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x="Aggregate rating",y="Ratings Counts",data=ratings)
# we made barplot by using x-axis and y-axis with ratings variable
# we also increase the size of the diagram by using (rcparams)
```

```python
sns.barplot(x="Aggregate rating",y="Ratings Counts",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','lightgreen','darkgreen'])
# in this we are giving the scale by using hue
# with the help of palette we are going to map the colour according to need (palette is used in bar ,count plots kinds)
```

**Observation**

1. Not Rated count is very high 
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','lightgreen','Darkgreen'])
# we are just count the rating variable with count plot 
# palette is used like above.
# we just use blue colour insted of useing white because it seems like invisible
```

**Observation Acoording to count plot**
1. white rating is minimum in count 
2. orange rating is high in count
3. both green are similar in ratings
